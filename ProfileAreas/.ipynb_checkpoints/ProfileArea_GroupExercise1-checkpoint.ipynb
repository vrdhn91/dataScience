Option 1: Handle the missing value by replacing it with 'missing_value' value

```python
imp_mean = SimpleImputer(missing_values='?', strategy='constant')
data_impute =imp_mean.fit_transform(data_csv)
```

```python
data_impute
```

Option 2: Handle the missing value by removing the record

```python
data_remove_empty = data_csv.drop(data_csv[data_csv['ca'] == '?'].index)
data_remove_empty = data_remove_empty.drop(data_remove_empty[data_remove_empty['thal'] == '?'].index)
```

```python
data_remove_empty.ca.unique()
data_remove_empty['ca'] = data_remove_empty['ca'].astype(float)
```

```python
data_remove_empty.thal.unique()
data_remove_empty['thal'] = data_remove_empty['thal'].astype(float)
```

# Exploratory Data Analysis

```python
import warnings
warnings.filterwarnings(action='ignore')
from matplotlib import pyplot as plt
import seaborn as sns
import pylab as pl
```

#### Univariate analysis: categorical data
The distribution of categorical variables in a dataset

```python
list_categorical = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal']
plt.figure(figsize=(18, 8))
plt.suptitle("Value counts of the categorical variables")
for i in range(len(list_categorical)):
    plt.subplot(2, 4, i+1)
    sns.countplot(data=data_remove_empty, x=list_categorical[i])
plt.show()
```

The distribution of categorical variables in a dataset, specifically with respect to the 'goal' variable.

```python
plt.figure(figsize=(18, 8))
plt.suptitle("Value counts of the categorical variables with respect to the \'goal\' varialbe")
for i in range(len(list_categorical)):
    plt.subplot(2, 4, i+1)
    sns.countplot(data=data_remove_empty, x=list_categorical[i], hue='goal')
plt.show()
```

#### Univerate analysis : numeric data
The distribution of numeric variables in a dataset

```python
data_remove_empty_numeric = data_remove_empty.drop(list_categorical, axis=1)
data_remove_empty_numeric.hist(bins=20, figsize=(9, 9))
pl.suptitle("Distribution of numeric variables")
plt.show()
```

#### Pairplots

Figuring out the relationship between two variables.

```python
sns.pairplot(data_remove_empty_numeric, hue='goal', height=2)
plt.show()
```

#### Boxplots
Distribution of a data with boxplots (numeric variables - 'goal')

```python
plt.figure(figsize = (18,20))
plt.suptitle("Distribution of a data with boxplots (numeric variables - 'goal')")
for i in range(6):
    plt.subplot(3,3,i + 1)
    sns.boxplot(x='goal', y=data_remove_empty_numeric.columns[i], data=data_remove_empty)
plt.show()
```

#### Scatter plots
Relationship between two specific variables. Pairs of variables were selected from the pairplot, appearing to have some correlation. And the scatter plots were created for an explict exploration.

age - trestbps  
age - chol

```python
plt.figure(figsize = (10,5))
plt.subplot(1, 2, 1)
sns.scatterplot(data=data_remove_empty, x='age', y='trestbps', hue='goal')
plt.legend(loc='upper left')
plt.subplot(1, 2, 2)
sns.scatterplot(data=data_remove_empty, x='age', y='chol', hue='goal')
plt.legend(loc='upper left')
plt.show()
```

Since it's not easily discernible visually, let's try to quantify it.
