```python
import matplotlib.pyplot as plt 
import matplotlib.image as img
import cv2
import numpy as np
```

```python
#elegans_plt = img.imread('elegans_001.tif') 
#plt.imshow(elegans_plt, cmap = plt.cm.gray)
img_elegans = cv2.imread('elegans_001.tif')
plt.imshow(img_elegans, cmap='gist_gray')
```

```python
gray_elegans = cv2.cvtColor(img_elegans, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_elegans, cmap='gist_gray')
```

```python
#Normalisation
cv2.normalize(gray_elegans, gray_elegans, 0, 255, cv2.NORM_MINMAX)
```

```python
plt.imshow(gray_elegans, cmap='gist_gray')
```

### Sobel Filter

```python
def sobel_filter_function(image):
    grad_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3, borderType=cv2.BORDER_DEFAULT)
    grad_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3, borderType=cv2.BORDER_DEFAULT)

    abs_grad_x = cv2.convertScaleAbs(grad_x)
    abs_grad_y = cv2.convertScaleAbs(grad_y)

    grad_sobel = cv2.addWeighted(abs_grad_x, 1, abs_grad_y, 1, 0)
    return grad_sobel
```

```python
grad_sobel = sobel_filter_function(gray_elegans)
```

```python
cv2.namedWindow('elegans_image_sobel',cv2.WINDOW_NORMAL)
cv2.resizeWindow('elegans_image_sobel', 1296,972)
cv2.imshow('elegans_image_sobel', grad_sobel)

cv2.waitKey(0)
```

### Gaussian Blur + Sobel Filter

```python
def blur_sobel_filter_function(image):
    blur_elegans = cv2.GaussianBlur(image, (3, 3), 0)
    
    grad_x = cv2.Sobel(blur_elegans, cv2.CV_64F, 1, 0, ksize=3, borderType=cv2.BORDER_DEFAULT)
    grad_y = cv2.Sobel(blur_elegans, cv2.CV_64F, 0, 1, ksize=3, borderType=cv2.BORDER_DEFAULT)

    abs_grad_x = cv2.convertScaleAbs(grad_x)
    abs_grad_y = cv2.convertScaleAbs(grad_y)

    grad_blur_sobel = cv2.addWeighted(abs_grad_x, 1, abs_grad_y, 1, 0)
    return grad_blur_sobel
```

```python
grad_blur_sobel = blur_sobel_filter_function(gray_elegans)
```

```python
cv2.namedWindow('elegans_image_blur_sobel',cv2.WINDOW_NORMAL)
cv2.resizeWindow('elegans_image_blur_sobel', 1296,972)
cv2.imshow('elegans_image_blur_sobel', grad_blur_sobel)

cv2.waitKey(0)
```

### Sharpen + Sobel Filter

```python
def sharpen_sobel_filter_function(image):
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    
    sharpened_elegans = cv2.filter2D(image, -1, kernel)
    
    grad_x = cv2.Sobel(sharpened_elegans, cv2.CV_64F, 1, 0, ksize=3, borderType=cv2.BORDER_DEFAULT)
    grad_y = cv2.Sobel(sharpened_elegans, cv2.CV_64F, 0, 1, ksize=3, borderType=cv2.BORDER_DEFAULT)

    abs_grad_x = cv2.convertScaleAbs(grad_x)
    abs_grad_y = cv2.convertScaleAbs(grad_y)

    grad_sharpen_sobel = cv2.addWeighted(abs_grad_x, 1, abs_grad_y, 1, 0)
    return grad_sharpen_sobel
```

```python
grad_sharpen_sobel = sharpen_sobel_filter_function(gray_elegans)
```

```python
cv2.namedWindow('elegans_image_sharpen_sobel',cv2.WINDOW_NORMAL)
cv2.resizeWindow('elegans_image_sharpen_sobel', 1296,972)
cv2.imshow('elegans_image_sharpen_sobel', grad_sharpen_sobel)

cv2.waitKey(0)
```

###  Scharr Filter

```python
def scharr_filter_function(image):
    grad_x = cv2.Scharr(image, cv2.CV_64F, 1, 0, borderType=cv2.BORDER_DEFAULT)
    grad_y = cv2.Scharr(image, cv2.CV_64F, 0, 1, borderType=cv2.BORDER_DEFAULT)

    abs_grad_x = cv2.convertScaleAbs(grad_x)
    abs_grad_y = cv2.convertScaleAbs(grad_y)

    grad_scharr = cv2.addWeighted(abs_grad_x, 1, abs_grad_y, 1, 0)
    return grad_scharr
```

```python
grad_scharr = scharr_filter_function(gray_elegans)
```

```python
cv2.namedWindow('elegans_image_scharr',cv2.WINDOW_NORMAL)
cv2.resizeWindow('elegans_image_scharr', 1296,972)
cv2.imshow('elegans_image_scharr', grad_scharr)

cv2.waitKey(0)
```

### Laplacian Filter

```python
def laplacian_filter_function(image):
    grad_x = cv2.Laplacian(image, cv2.CV_64F, ksize=3, borderType=cv2.BORDER_DEFAULT)
    grad_y = cv2.Laplacian(image, cv2.CV_64F, ksize=3, borderType=cv2.BORDER_DEFAULT)

    abs_grad_x = cv2.convertScaleAbs(grad_x)
    abs_grad_y = cv2.convertScaleAbs(grad_y)

    grad_laplacian = cv2.addWeighted(abs_grad_x, 1, abs_grad_y, 1, 0)
    return grad_laplacian
```

```python
grad_laplacian = laplacian_filter_function(gray_elegans)
```

```python
cv2.namedWindow('elegans_image_laplacian',cv2.WINDOW_NORMAL)
cv2.resizeWindow('elegans_image_laplacian', 1296,972)
cv2.imshow('elegans_image_laplacian', grad_laplacian)

cv2.waitKey(0)
```
